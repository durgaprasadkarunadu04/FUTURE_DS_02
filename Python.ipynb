```python
import pandas as pd

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head()
```

```python
print("Missing values:")
print(df.isnull().sum())
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df = df.dropna(subset=['TotalCharges'])
```

```python
print("\nDuplicates:", df.duplicated().sum())
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].map({1: 'Senior', 0: 'Non-Senior'})

print("\nCleaned shape:", df.shape)
df.head()
```

```python
def tenure_group(tenure):
    if tenure <= 12:
        return '0-12 Months'
    elif tenure <= 24:
        return '13-24 Months'
    elif tenure <= 48:
        return '25-48 Months'
    else:
        return '49+ Months'

df['Tenure_Group'] = df['tenure'].apply(tenure_group)

print("Tenure groups added:")
print(df['Tenure_Group'].value_counts())
```

```python
total = len(df)
churned = df[df['Churn'] == 'Yes'].shape[0]
retained = df[df['Churn'] == 'No'].shape[0]
churn_rate = round(churned / total * 100, 2)

print(f"Total Customers: {total}")
print(f"Churned: {churned}")
print(f"Retained: {retained}")
print(f"Churn Rate: {churn_rate}%")
```

```python
contract_churn = df.groupby(['Contract', 'Churn']).size().reset_index(name='Count')
print("Churn by Contract Type:")
print(contract_churn)
```

```python
tenure_churn = df.groupby(['Tenure_Group', 'Churn']).size().reset_index(name='Count')
print("Churn by Tenure Group:")
print(tenure_churn)
```

```python
internet_churn = df.groupby(['InternetService', 'Churn']).size().reset_index(name='Count')
print("Churn by Internet Service:")
print(internet_churn)
```

```python
payment_churn = df.groupby(['PaymentMethod', 'Churn']).size().reset_index(name='Count')
print("Churn by Payment Method:")
print(payment_churn)
```

```python
avg_charges = df.groupby('Churn')['MonthlyCharges'].mean().reset_index()
avg_charges.columns = ['Churn', 'Avg_Monthly_Charges']
avg_charges['Avg_Monthly_Charges'] = avg_charges['Avg_Monthly_Charges'].round(2)
print("Avg Monthly Charges by Churn:")
print(avg_charges)
```

```python
senior_churn = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
print("Churn by Senior Citizen:")
print(senior_churn)
```

```python
tableau_columns = [
    'customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'tenure', 'Tenure_Group', 'PhoneService', 'InternetService',
    'Contract', 'PaperlessBilling', 'PaymentMethod',
    'MonthlyCharges', 'TotalCharges', 'Churn'
]

df_tableau = df[tableau_columns]
df_tableau.to_csv('churn_cleaned.csv', index=False)

print(f"✅ churn_cleaned.csv saved!")
print(f"Shape: {df_tableau.shape}")
df_tableau.head()
```
